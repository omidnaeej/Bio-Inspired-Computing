### **Traveling Salesman Problem using Genetic Algorithm**


#### **Fitness Function** 
The distance traveled by the Salesman :)

```python
class Chromosome:
    def __init__(self, size:int) -> None:
        self.size = size
        self.permutation = list(range(size))
        random.shuffle(self.permutation)
        self.fitness = size
        self.calc_fitness()
    
    def mutation(self, alpha= .9) -> None:
        if random.random() < alpha:
            a = random.randint(0,self.size-1)
            b = random.randint(0,self.size-1)
            self.permutation[a], self.permutation[b] = self.permutation[b], self.permutation[a]
        self.calc_fitness()
    
    def calc_fitness(self) -> int:
        res = 0
        for i in range(self.size-1):
            x = self.permutation[i]
            y = self.permutation[i+1]
            distance = math.sqrt((coordinates[x][0] - coordinates[y][0])**2 
                                 + (coordinates[x][1] - coordinates[y][1])**2)
            res += distance
        self.fitness = res
```

#### **How Crossover Works?**


```python
def cross_over(par1 : Chromosome, par2:Chromosome) -> Chromosome:
    children = Chromosome(par1.size)
    boundes = list(random.sample(range(par1.size), 2))
    boundes.sort() 
    start, end = boundes[0], boundes[1]
    
    center = par2.permutation[start : end+1]  
    left = []
    right = []
    counter = 0

    for i in range(par1.size):
        if (par1.permutation[i] not in center):
            if (counter < start):
                left.append(par1.permutation[i])
            else:
                right.append(par1.permutation[i])
            counter += 1
            
    children.permutation = left + center + right
    children.calc_fitness()
    return children
```

#### **Selection ?**
- *`Truncation Selection`* = sort the population based on *fitness* and then randomly select some members from among the top percentage members of the population.

#### **Replacement ?**
- *`Steady State Replacement`* = Keeping the majority of the parent population and replacing the best offspring produced with part of the parent

```python
def GA(pop_size, Chromosome_size, iteration= 8000) -> Chromosome:
    population = [Chromosome(Chromosome_size) for _ in range(pop_size)]
    population.sort(key = lambda x :x.fitness)
    
    count = 0
    perv = 0
    step = 0
    
    while (True):
        children = []
        sample = population[:int(pop_size*.2)]
        for _ in range(pop_size):
            parents = random.choices(sample, k = 2)
            child = cross_over(*parents)
            child.mutation()
            children.append(child)

            child2 = (cross_over(parents[1], parents[0]))
            child2.mutation()
            children.append(child2)
                    
        children.sort(key = lambda x :x.fitness)
        
        population = population[ :pop_size-1] + children[ :1] 
        """ trade-off between exploration & exploitation """
        population.sort(key = lambda x :x.fitness)

        print('best fitness: ' , round(population[0].fitness, 3), "\t \t worse fitness: " , round(population[-1].fitness, 3) ,sep='')
        
        if (population[0].fitness == perv):
            count +=1
        else:
            count = 0
            perv = population[0].fitness
        
        if (count > 30 and step>iteration) or count>100 or step>8000:
            break
        
        step +=1
        
    return population[0]      
```
